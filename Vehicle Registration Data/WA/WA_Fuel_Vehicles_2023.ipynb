```python
# Merge df2 with df3 to get the 'Vehicle Category' based on 'Record Type', 'Make', and 'Body Type'.
df4 = pd.merge(df2, df3, on=['Make', 'Model', 'Vehicle Type'], how='left')

# Select the required columns for the new dataframe df5
df4 = df4[['Vehicle Type NEW', 'State', 'Postal Code', 'County', 'Model Year', 'Fuel Type Primary', 'Transaction Month and Year']]

# Save df5 to a CSV file
#df4.to_csv('df5.csv', index=False)

# Check the first few rows of the new dataframe
print(df4.head(100))
```

```python
df4.head(100)
```

```python
df4['Model Year'].unique()
```

```python
df4['Fuel Type Primary'].unique()
```

```python
# Copy df4 to df5 to retain the original data (correcting from df5 = df4.copy() to the correct form)
df5 = df4.copy()

# Renaming columns
df5.rename(columns={'Vehicle Type NEW': 'Vehicle Type', 'Fuel Type Primary': 'Fuel Type', 'Postal Code': 'Zip'}, inplace=True)

# Calculate 'Age' from 'Model Year', where 2024 - Model Year. Cap the age at 30 for vehicles older than 1994.
# Assuming 'Model Year' exists in df5, correcting df5['Age'] calculation based on its own 'Model Year' column.
df5['Age'] = 2024 - df5['Model Year'].astype(int)
df5['Age'] = df5['Age'].apply(lambda x: x if x <= 30 else 30)

# Drop the 'Model Year' column as it's replaced by 'Age'
df5.drop('Model Year', axis=1, inplace=True)

# Mapping for fuel types
fuel_type_mapping = {
    'Gasoline': 'Gasoline',
    'Electric': 'Electricity',
    'Flexible Fuel Vehicle (FFV)': 'Ethanol(E-85)',
    'Diesel': 'Diesel',
    'Ethanol (E85)': 'Ethanol(E-85)',
    'Not Applicable': 'Other',
    'Compressed Natural Gas (CNG)': 'Compressed natural Gas (CNG)',
    'Liquefied Petroleum Gas (propane or LPG)': 'Compressed natural Gas (CNG)',
    'Hybrid': 'Electricity',
    'Other': 'Other',
    'Steam': 'Other',
    'Butane': 'Other',
    'Hydrogen Fuel Cell': 'Electricity',
    'Natural Gas': 'Compressed natural Gas (CNG)'
}

# Apply the mapping to the 'Fuel Type' column
df5['Fuel Type'] = df5['Fuel Type'].map(fuel_type_mapping)

# Add 'Vehicle Count' column initialized to 1
df5['Vehicle Count'] = 1

# Filter out rows where 'Fuel Type' is 'Other'
df5 = df5[df5['Fuel Type'] != 'Other']

# Save the dataframe to a CSV file
#df5.to_csv('df6.csv', index=False)

# Display the first few rows of the new dataframe
print(df5.head(100))
```

```python
df5.head(100)
```

```python
num_rows = df5.shape[0]
print("Number of rows:", num_rows)
```

```python
# Group by 'Vehicle Type', 'Fuel Type', and 'Age'
df6 = df5.groupby(['Vehicle Type', 'Fuel Type', 'Age'])['Vehicle Count'].sum().reset_index()

# Save df6 to a CSV file
#df6.to_csv('df6.csv', index=False)

# Display the first few rows of the new dataframe
print(df6.head(100))
```
