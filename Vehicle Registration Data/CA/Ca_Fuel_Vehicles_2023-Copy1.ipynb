```python
# Creating df2 by dropping rows based on specified conditions
df2 = df1.drop(df1[(df1['Make'] == 'OTHER/UNK') | 
                   (df1['Duty'] == 'Unk') | 
                   (df1['Fuel'].isin(['Unk', 'Other']))].index)

# Displaying first 50 rows of df2
print(df2.head(50))
```

```python
# Convert the "Date" column to datetime format
df2['Date'] = pd.to_datetime(df2['Date'])

# Extract the year part from the "Date" column
df2['Year'] = df2['Date'].dt.year

# Find unique years
unique_years = df2['Year'].unique()

# Display the unique years
print(unique_years)
```

```python
# Merge df3 with df4 to get the 'Vehicle Category' based on 'Record Type', 'Make', and 'Body Type'.
df4 = pd.merge(df2, df3, on=['Make', 'Duty'], how='left')

# Select the required columns for the new dataframe df5
df4 = df4[['Vehicle Type','Zip Code','Model Year', 'Fuel', 'Date', 'Vehicles']]

# Save df4 to a CSV file
#df4.to_csv('df4.csv', index=False)

# Check the first few rows of the new dataframe
print(df4.head(100))
```

```python
num_rows = df4.shape[0]
print("Number of rows:", num_rows)
```

```python
df4['Model Year'].unique()
```

```python
# Copy df5 to df6 to retain the original data
df5 = df4.copy()

# Renaming 'Vehicle Category' to 'Vehicle Type'
df5.rename(columns={'Zip Code': 'Zip','Fuel': 'Fuel Type','Vehicles': 'Vehicle Count'}, inplace=True)
df5.rename(columns={'Fuel': 'Fuel Type'}, inplace=True)

# Custom function to handle special cases in 'Model Year'
def convert_model_year_to_int(model_year):
    if model_year == '<2009':
        return 2008  # Or another appropriate value
    else:
        return int(model_year)

# Apply the custom function to the 'Model Year' column
df5['Model Year'] = df5['Model Year'].apply(convert_model_year_to_int)

# Now calculate 'Age'
df5['Age'] = 2024 - df5['Model Year']
df5['Age'] = df5['Age'].apply(lambda x: x if x <= 30 else 30)

# Drop the 'Model Year' column as it's replaced by 'Age'
df5.drop('Model Year', axis=1, inplace=True)


# Mapping for fuel types
fuel_type_mapping = {
    'Gasoline': 'Gasoline',
    'Diesel and Diesel Hybrid':'Diesel',
    'Hybrid Gasoline':'Gasoline',
    'Flex-Fuel':'Ethanol (E-85)',
    'Battery Electric':'Electricity',
    'Plug-in Hybrid':'Electricity',
    'Natural Gas':'Compressed natural Gas (CNG)',
    'Hydrogen Fuel Cell': 'Electricity'
}

# Apply the mapping to the 'Fuel Type' column
df5['Fuel Type'] = df5['Fuel Type'].map(fuel_type_mapping)

# Add 'Vehicle Count' column initialized to 1
#df5['Vehicle Count'] = 1

# Save df6 to a CSV file
#df5.to_csv('df5.csv', index=False)

# Display the first few rows of the new dataframe
print(df5.head(100))
```

```python
df5['Age'].unique()
```

```python
# Group by 'Vehicle Type', 'Fuel Type', and 'Age'
df6 = df5.groupby(['Vehicle Type', 'Fuel Type', 'Age'])['Vehicle Count'].sum().reset_index()

# Save df6 to a CSV file
#df6.to_csv('df7.csv', index=False)

# Display the first few rows of the new dataframe
print(df6.head(100))
```
