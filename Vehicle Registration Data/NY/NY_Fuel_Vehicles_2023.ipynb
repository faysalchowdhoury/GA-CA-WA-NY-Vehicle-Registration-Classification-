```python
# Merge df3 with df4 to get the 'Vehicle Category' based on 'Record Type', 'Make', and 'Body Type'.
df5 = pd.merge(df3, df4, on=['Record Type', 'Make', 'Body Type'], how='left')

# Select the required columns for the new dataframe df5
df5 = df5[['Vehicle Category', 'State', 'Zip', 'County', 'Model Year', 'Fuel Type', 'Reg Valid Date', 'Reg Expiration Date']]

# Save df5 to a CSV file
#df5.to_csv('df5.csv', index=False)

# Check the first few rows of the new dataframe
print(df5.head(100))
```

```python
# Copy df5 to df6 to retain the original data
df6 = df5.copy()

# Renaming 'Vehicle Category' to 'Vehicle Type'
df6.rename(columns={'Vehicle Category': 'Vehicle Type'}, inplace=True)

# Calculate 'Age' from 'Model Year', where 2024 - Model Year. Cap the age at 30 for vehicles older than 1994.
df6['Age'] = 2024 - df6['Model Year'].astype(int)
df6['Age'] = df6['Age'].apply(lambda x: x if x <= 30 else 30)

# Drop the 'Model Year' column as it's replaced by 'Age'
df6.drop('Model Year', axis=1, inplace=True)

# Mapping for fuel types
fuel_type_mapping = {
    'GAS': 'Gasoline',
    'PROPANE': 'Compressed natural Gas (CNG)', # Assuming propane as a form of compressed natural gas
    'DIESEL': 'Diesel',
    'FLEX': 'Ethanol(E-85)', # Assuming flex fuel is primarily ethanol
    'ELECTRIC': 'Electricity',
    'COMP N/G': 'Compressed natural Gas (CNG)' # Assuming 'COMP N/G' stands for compressed natural gas
}

# Apply the mapping to the 'Fuel Type' column
df6['Fuel Type'] = df6['Fuel Type'].map(fuel_type_mapping)

# Add 'Vehicle Count' column initialized to 1
df6['Vehicle Count'] = 1

# Save df6 to a CSV file
#df6.to_csv('df6.csv', index=False)

# Display the first few rows of the new dataframe
print(df6.head(100))
```
