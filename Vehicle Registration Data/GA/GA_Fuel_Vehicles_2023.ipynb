```python
# Function to extract desired text from the Vehicle Type column
def extract_text(text):
    # Define a list of desired vehicle types
    desired_types = [
        'Combination Long-haul Truck',
        'Combination Short-haul Truck',
        'Light Commercial Truck',
        'Motor Home',
        'Motorcycle',
        'Other Buses',
        'Passenger Car',
        'Passenger Truck',
        'Refuse Truck',
        'School Bus',
        'Single Unit Long-haul Truck',
        'Single Unit Short-haul Truck',
        'Transit Bus'
    ]
    # Iterate through the desired types and return if found in the text
    for vehicle_type in desired_types:
        if vehicle_type in text:
            return vehicle_type
    # If none of the desired types are found, return the original text
    return text

# Apply the function to the "Vehicle Type" column
df5['Vehicle Type'] = df5['Vehicle Type'].apply(extract_text)

# Display the first few rows of the modified DataFrame
print(df5.head())
```

```python
# Merge df3 with df4 to get the 'Vehicle Category' based on 'Record Type', 'Make', and 'Body Type'.
df6 = pd.merge(df3, df5, on=['Make', 'Duty'], how='left')

# Select the required columns for the new dataframe df5
df6 = df6[['Vehicle Type', 'year_model', 'Fuel Detail', 'COUNT(1)']]

# Save df6 to a CSV file
#df6.to_csv('df6.csv', index=False)

# Check the first few rows of the new dataframe
print(df6.head(100))
```

```python
# Copy df4 to df5 to retain the original data (correcting from df5 = df4.copy() to the correct form)
df7 = df6.copy()

# Renaming columns
df7.rename(columns={'year_model':'Model Year','Fuel Detail': 'Fuel Type', 'COUNT(1)': 'Vehicle Count'}, inplace=True)

# Calculate 'Age' from 'Model Year', where 2024 - Model Year. Cap the age at 30 for vehicles older than 1994.
# Assuming 'Model Year' exists in df5, correcting df5['Age'] calculation based on its own 'Model Year' column.
df7['Age'] = 2024 - df7['Model Year'].astype(int)
df7['Age'] = df7['Age'].apply(lambda x: x if x <= 30 else 30)

# Drop the 'Model Year' column as it's replaced by 'Age'
df7.drop('Model Year', axis=1, inplace=True)

# Mapping for fuel types
fuel_type_mapping = {
        'Unknown': 'Unknown',  # Decide how to handle unknown values
        'Diesel': 'Diesel',
         'CONVERTIBLE': 'Gasoline',
        'Diesel-Electric Hybrid': 'Diesel',
        'Gasoline': 'Gasoline',
        'GAS': 'Gasoline',
        'E-85': 'Ethanol(E-85)',
        'FLEXIBLE': 'Ethanol(E-85)',
        'CNG': 'Compressed natural Gas (CNG)',
        'COMPRESSED NATURAL GAS': 'Compressed natural Gas (CNG)',
        'Electric': 'Electricity',
        'HYDROGEN FUEL CELL': 'Electricity',
        'Propane': 'Compressed natural Gas (CNG)',
        'Gasoline-Electric Hybrid': 'Gasoline',
        'ELECTRIC AND GAS HYBRID': 'Gasoline',
        'ELECTRIC AND GAS HYBRID PHEV': 'Gasoline',
}

# Apply the mapping to the 'Fuel Type' column
df7['Fuel Type'] = df7['Fuel Type'].map(fuel_type_mapping)

# Add 'Vehicle Count' column initialized to 1
df7['Vehicle Count'] = 1

# Filter out rows where 'Fuel Type' is 'Other'
df7 = df7[df7['Fuel Type'] != 'Unknown']

# Save the dataframe to a CSV file
#df7.to_csv('df7.csv', index=False)

# Display the first few rows of the new dataframe
print(df7.head(100))
```

```python
# Group by 'Vehicle Type', 'Fuel Type', and 'Age'
df8 = df7.groupby(['Vehicle Type', 'Fuel Type', 'Age'])['Vehicle Count'].sum().reset_index()

# Save df7 to a CSV file
#df7.to_csv('df7.csv', index=False)

# Display the first few rows of the new dataframe
print(df8.head())
```
